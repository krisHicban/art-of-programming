### Exercițiul 1 – Mini inventar Pandas pornit din date NumPy
În continuarea array-urilor sintetice din recapitularea `13-14_NumPY`, construiește un `DataFrame` Pandas cu coloanele `Produs`, `Preț` și `Stoc`, folosind date generate de tine (ideal structurate cu liste sau array-uri). Asigură-te că tabela are cel puțin cinci rânduri și inspectează rezultatul pentru a confirma schema.


### Exercițiul 2 – Selectează rapid ofertele premium
Folosește DataFrame-ul din exercițiul anterior pentru a extrage doar produsele cu `Preț` mai mare de 50. Compare filtrarea cu mască booleană cu o abordare `loc` sau `.query()`, astfel încât să vezi diferențele de sintaxă și expresivitate.


### Exercițiul 3 – Pregătește ordinea pentru analize ulterioare
Continuăm pregătirea datasetului pentru vizualizările din modulul 18-19. Sortează produsele descrescător după `Preț` cu `sort_values()` și afișează primele trei intrări pentru a verifica rezultatul.


### Exercițiul 4 – Calculează automat TVA-ul
Adaugă o coloană `TVA` ce reprezintă 19% din prețul fiecărui produs și păstrează două zecimale prin rotunjire vectorizată. Observă cum Pandas aplică operațiile pe tot coloana și confirmă cu `head()` că valorile se potrivesc așteptărilor.
